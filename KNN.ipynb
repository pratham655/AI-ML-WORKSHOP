```python
x=data.drop(['Id','Species'],axis=1)
```

```python
x.shape
```

```python
x.head()
```

```python
y=data['Species']
# 2 brackets means data frame ([]),1 means 1D it has only one value that is species in this case
```

```python
y.value_counts()
#to check value counts it should be either panda series or dataframe
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=3)
#for splitting the training and testing model by 80 and 20 percent respectively
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
# to check in the y train model by splitting each species into 80% for the training and rest 20% for testing respectively and this will store as panda 1d series
ytrain.value_counts()
#to get exact 80% in the y model u have to change the random_state to some perfect value to get a perfect 80% value
#here we have taken 3 as random state to get 40 each for each species
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
#so now we are getting the model ready by putting xtrain and ytrain and now we are storing in the new variable ypred which gives model output by testing with xtest and the actual output which is ytest
```

```python
ypred
```

```python
ytest.values
#ytest is a panda series so we need to convert it into array and now we'll get the results
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
ac=accuracy_score(ytest,ypred)
ac
```

```python
cm=confusion_matrix(ytest,ypred)
cm
#we found confusion matrix
```
