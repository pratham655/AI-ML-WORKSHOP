```python
import nltk
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
sia=SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores("THE MOVIE IS VERY GOOD")
```

```python
#vader is a sentiment analysis tool design to understand the emotional tone of the text for eg tweets,reviews
#lexicon is a sentiment dictionary where it tests whether is it positive or negative
#sentimentintensityanalyser is an object inside the vader
#compound means average fo negative and positive -1 means -ve and +1 means +ve
```
