```python
import pandas as pd
import nltk
```

```python
text="""Python is a widely used programming language because it is easy to learn and works well for many fields like web development, data science, and automation. As a result, Python programmers are in high demand, with many companies looking for people who can use it to solve real-world problems."""
```

```python
from nltk.tokenize import word_tokenize
```

```python
nltk.download("punkt")
```

```python
nltk.download("punkt_tab")
```

```python
word_final=word_tokenize(text)
```

```python
word_final
```

```python
len(word_final)
```

```python
from nltk.corpus import stopwords
nltk.download("stopwords")
stopwords.words("english")
final_word=[]
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
stem_word=[]
for i in word_final:
    if i not in stopwords.words("english"):
       final_word.append(i)
       stem_word.append(stemmer.stem(i))
```

```python
len(final_word)
```

```python
df=pd.DataFrame()
df['fianl_word']=final_word
df['stem_word']=stem_word
df
```

```python
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
nltk.download("wordnet")
final_words=[]
lemma_word=[]
for i in word_final:
  if i not in stopwords.words("english"):
    final_words.append(i)
    lemma_word.append(lemma.lemmatize(i,pos="v"))
```

```python
  df['Lemma']=lemma_word
```
