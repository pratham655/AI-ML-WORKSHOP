```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/titanic.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
cols=['PassengerId','Name','Ticket','Fare','Cabin']
data=data.drop(cols,axis=1)
```

```python
data.head()
```

```python
data['Age'].isnull().sum()
```

```python
mean_age=round(data['Age'].mean())
```

```python
mean_age
```

```python
data['Age']=data['Age'].fillna(mean_age)
```

```python
data['Age']
```

```python
data.info()
```

```python
data=data.dropna(axis=0)
```

```python
data.info()
```

```python
cols=['Sex','Embarked']
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
data['Sex']=enc.fit_transform(data['Sex'])
data['Embarked']=enc.fit_transform(data['Embarked'])
```

```python
data.info()
```

```python
data['Age']=data['Age'].astype(int)
```

```python
data.info()
```

```python
x=data.drop(['Survived'],axis=1)
```

```python
x.shape
```

```python
y=data['Survived']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.85,random_state=0)
```

```python
y.value_counts()
```

```python
pd.DataFrame(ytrain).value_counts()
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
ypred
```

```python
ytest.values
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
ac=accuracy_score(ytest,ypred)
```

```python
ac
```

```python
cm=confusion_matrix(ytest,ypred)
```

```python
cm
```
