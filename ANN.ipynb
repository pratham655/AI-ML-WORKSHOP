```python
#classification of data using ann in deep learning
import tensorflow
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import*
#tensorflow is a builtin app which is in colab where keras is a model which contains all the features
```

```python
data=pd.read_csv('/content/breast_cancer _ day1.csv')
```

```python
data=data.drop(['id'],axis=1)
```

```python
x=data.drop(['diagnosis'],axis=1)
#we removed id and it became 31 rows
#x axis should have all the features while y should have binary result whether cancer is there or not
```

```python
y=data['diagnosis']
```

```python
y.shape
#here we have diagnosis from the x row and kept in y to get result whether person should be diagonised or not
```

```python
#we use standard scaler function because we have different data types thats why
#and also we are converting the label encounter to integer
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
y=pd.DataFrame(y)

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x
```

```python
x=pd.DataFrame(x)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import(train_test_split)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
#y is panda series binary
y.value_counts()
#not the total 80% of 569 we need 80% of both b and m values respectively so that the split what we do should be correct
```

```python
ytrain.value_counts()
```

```python
model=Sequential()
```

```python
model.add(Dense(16,input_dim=30,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(xtrain,ytrain,epochs=50,validation_data=(xtest,ytest))
```

```python
ypred=model.predict(xtest)
```

```python
ypred
ypred=ypred.round()
```
